```python
import pandas as pd
import matplotlib.pyplot as plt

# Percorso del file di input
input_file = "baseline.csv"
output_file = "output.csv"

# Legge il CSV
df = pd.read_csv(input_file)

# Aggiunge la colonna 'tmp' con valori da 1 a n
df["tmp"] = range(1, len(df) + 1)

# Salva il nuovo file
df.to_csv(output_file, index=False)

print(f"File salvato con successo in: {output_file}")
```

```python
# === Parametri ===
input_file = "output.csv"
max_rows = 15000

df_plot = df.head(max_rows)

plt.figure(figsize=(10, 5))

plt.plot(df_plot["tmp"], df_plot[df.columns[0]], label=df_plot.columns[0])
plt.plot(df_plot["tmp"], df_plot[df.columns[3]], label=df_plot.columns[3])
plt.plot(df_plot["tmp"], df_plot[df.columns[9]], label=df_plot.columns[9])

plt.xlabel("Tempo (tmp)")
plt.ylabel("Registri")
plt.title(f"Prime {max_rows} righe di {input_file}")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
